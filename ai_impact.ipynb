```python
# ============================================
# üîé EDA Inicial - Global AI Content Impact Dataset
# ============================================

# 1. Importar librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo equivalente a seaborn
plt.style.use("seaborn-v0_8")

# Paleta de colores para seaborn
sns.set_palette("viridis")


# 2. Cargar el dataset
file_path = "Global_AI_Content_Impact_Dataset.csv"  # Ajusta la ruta si es necesario
df = pd.read_csv(file_path)

# 3. Informaci√≥n b√°sica
print("üìä Dimensiones del dataset:", df.shape)
print("\nüßæ Tipos de datos:\n", df.dtypes)
print("\nüîç Valores nulos por columna:\n", df.isnull().sum())

# 4. Vista previa de datos
display(df.head())

# 5. Resumen estad√≠stico
print("\nüìà Resumen estad√≠stico:")
display(df.describe())

# 6. Variables categ√≥ricas - distribuci√≥n de frecuencias
categorical_cols = df.select_dtypes(include=["object"]).columns
for col in categorical_cols:
    print(f"\nüìä Distribuci√≥n de {col}:")
    print(df[col].value_counts())
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f"Distribuci√≥n de {col}")
    plt.show()

# 7. Histogramas de variables num√©ricas
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols].hist(figsize=(15,12), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("Distribuci√≥n de variables num√©ricas", fontsize=16)
plt.show()

# 8. Boxplots para detecci√≥n de outliers
plt.figure(figsize=(15,8))
sns.boxplot(data=df[numeric_cols])
plt.xticks(rotation=45)
plt.title("Distribuci√≥n y outliers en variables num√©ricas")
plt.show()

# 9. Mapa de correlaciones
plt.figure(figsize=(12,8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Mapa de correlaciones entre variables num√©ricas")
plt.show()

# ============================================
# ‚úÖ Fin del EDA inicial
# ============================================
```
