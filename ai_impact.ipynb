```python
# ============================================
# üîé EDA Inicial - Global AI Content Impact Dataset
# Paleta violeta-azulada
# ============================================

# 1. Importar librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n de visualizaci√≥n (violetas y azules)
plt.style.use("seaborn-v0_8-darkgrid")
sns.set_palette("mako")  # tonos violeta-azulados
pd.set_option("display.max_columns", None)

# 2. Cargar el dataset
file_path = "Global_AI_Content_Impact_Dataset.csv"  # Ajusta la ruta si es necesario
df = pd.read_csv(file_path)

# 3. Informaci√≥n b√°sica
print("üìä Dimensiones del dataset:", df.shape)
print("\nüßæ Tipos de datos:\n", df.dtypes)
print("\nüîç Valores nulos por columna:\n", df.isnull().sum())

# 4. Vista previa de datos
display(df.head())

# 5. Resumen estad√≠stico
print("\nüìà Resumen estad√≠stico:")
display(df.describe())

# 6. Variables categ√≥ricas - distribuci√≥n de frecuencias
categorical_cols = df.select_dtypes(include=["object"]).columns
for col in categorical_cols:
    print(f"\nüìä Distribuci√≥n de {col}:")
    print(df[col].value_counts())
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette="crest")
    plt.xticks(rotation=45, ha="right")
    plt.title(f"Distribuci√≥n de {col}", fontsize=12, color="navy")
    plt.show()

# 7. Histogramas de variables num√©ricas
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols].hist(figsize=(15,12), bins=15, color="#5D3FD3", edgecolor='black')
plt.suptitle("Distribuci√≥n de variables num√©ricas", fontsize=16, color="darkblue")
plt.show()

# 8. Boxplots para detecci√≥n de outliers
plt.figure(figsize=(15,8))
sns.boxplot(data=df[numeric_cols], palette="mako")
plt.xticks(rotation=45)
plt.title("Distribuci√≥n y outliers en variables num√©ricas", fontsize=14, color="midnightblue")
plt.show()

# 9. Mapa de correlaciones
plt.figure(figsize=(12,8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap="PuBuGn", fmt=".2f", linewidths=0.5, cbar_kws={'shrink': 0.8})
plt.title("Mapa de correlaciones entre variables num√©ricas", fontsize=14, color="darkblue")
plt.show()

# ============================================
# ‚úÖ Fin del EDA inicial
# ============================================
```

```python
# ============================================
# üîÑ ETL - Global AI Content Impact Dataset
# ============================================

# 1. Importar librer√≠as
import pandas as pd

# 2. EXTRACTION - Cargar datos
file_path = "Global_AI_Content_Impact_Dataset.csv"  # Ajusta la ruta si es necesario
df_raw = pd.read_csv(file_path)

print("‚úÖ Datos extra√≠dos")
print("Dimensiones iniciales:", df_raw.shape)

# 3. TRANSFORMATION - Limpieza y preparaci√≥n
df = df_raw.copy()

# --- A) Normalizaci√≥n de nombres de columnas ---
df.columns = (df.columns
              .str.strip()
              .str.replace(" ", "_")
              .str.replace("-", "_")
              .str.replace(r"[%()]", "", regex=True)
              .str.lower()
              .str.replace("__", "_"))  # corregir dobles guiones bajos

print("\nüìë Nuevos nombres de columnas:")
print(df.columns.tolist())

# --- B) Eliminar duplicados ---
duplicates = df.duplicated().sum()
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"\n‚ö†Ô∏è {duplicates} duplicados eliminados")
else:
    print("\n‚úÖ No se encontraron duplicados")

# --- C) Revisar valores nulos ---
print("\nüîç Valores nulos por columna:")
print(df.isnull().sum())

# --- D) Conversi√≥n de tipos ---
df["year"] = df["year"].astype(int)
numeric_cols = df.select_dtypes(include="number").columns
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

# --- E) Normalizar categor√≠as ---
if "regulation_status" in df.columns:
    df["regulation_status"] = df["regulation_status"].str.strip().str.capitalize()

# --- F) Asegurar nombres clave para m√©tricas ---
rename_map = {
    "ai_adoption_rate_": "ai_adoption_rate",
    "ai_adoption_rate__": "ai_adoption_rate",
    "human_ai_collaboration_rate": "human_ai_collaboration_rate",
    "job_loss_due_to_ai": "job_loss_due_to_ai",
    "revenue_increase_due_to_ai": "revenue_increase_due_to_ai"
}
df = df.rename(columns={k: v for k, v in rename_map.items() if k in df.columns})

print("\nüìë Columnas finales despu√©s de renombrado:")
print(df.columns.tolist())

# --- G) Crear columna derivada ---
if {"ai_adoption_rate", "revenue_increase_due_to_ai", "human_ai_collaboration_rate", "job_loss_due_to_ai"}.issubset(df.columns):
    df["ai_impact_index"] = (
        df["ai_adoption_rate"]*0.4 +
        df["revenue_increase_due_to_ai"]*0.3 +
        df["human_ai_collaboration_rate"]*0.2 -
        df["job_loss_due_to_ai"]*0.1
    )
    print("\nüÜï Columna derivada 'ai_impact_index' creada con √©xito")
else:
    print("\n‚ö†Ô∏è No se encontraron todas las columnas necesarias para calcular 'ai_impact_index'")

# 4. LOAD - Guardar dataset limpio
output_file = "Global_AI_Content_Clean.csv"
df.to_csv(output_file, index=False)

print(f"\nüíæ Dataset limpio guardado como: {output_file}")
print("Dimensiones finales:", df.shape)

# Vista previa final
display(df.head())
```

```python
# ============================================
# üìä EDA Final Avanzado - Global AI Dataset
# ============================================

# 1. Importar librer√≠as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from scipy.stats import f_oneway

# Estilo
plt.style.use("seaborn-v0_8-darkgrid")
sns.set_palette("mako")
pd.set_option("display.max_columns", None)

# 2. Cargar dataset limpio
file_path = "Global_AI_Content_Clean.csv"
df = pd.read_csv(file_path)

print("‚úÖ Dataset cargado")
print("Dimensiones:", df.shape)

# ============================================
# üîπ 1. Exploraci√≥n de columnas
# ============================================
print("\nüìã Columnas disponibles:")
print(df.columns.tolist())

# Separar categ√≥ricas y num√©ricas
categorical_cols = df.select_dtypes(include=["object"]).columns
numeric_cols = df.select_dtypes(include=np.number).columns

print("\nüî¢ Variables num√©ricas:", list(numeric_cols))
print("\nüî† Variables categ√≥ricas:", list(categorical_cols))

# ============================================
# üîπ 2. Distribuciones KDE
# ============================================
plt.figure(figsize=(12,8))
for col in numeric_cols[:5]:  # primeras 5 variables num√©ricas
    sns.kdeplot(df[col].dropna(), shade=True, label=col)
plt.title("Distribuciones KDE de variables num√©ricas", fontsize=14, color="darkviolet")
plt.legend()
plt.show()

# ============================================
# üîπ 3. Pairplot Multivariado
# ============================================
sns.pairplot(df[numeric_cols[:4]], corner=True, diag_kind="kde", palette="mako")
plt.suptitle("Pairplot entre variables clave", y=1.02, fontsize=14, color="navy")
plt.show()

# ============================================
# üîπ 4. Clustering KMeans
# ============================================
X = df[numeric_cols].dropna()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["cluster"] = kmeans.fit_predict(X_scaled)

# Usar 2 primeras variables num√©ricas disponibles para el scatter
x_var, y_var = numeric_cols[:2]

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x=x_var, y=y_var,
                hue="cluster", palette="viridis", alpha=0.8)
plt.title(f"Clustering seg√∫n {x_var} vs {y_var}", fontsize=13, color="darkblue")
plt.show()

# ============================================
# üîπ 5. ANOVA: Diferencias por regulaci√≥n (si existe)
# ============================================
if "regulation_status" in df.columns:
    groups = [group[numeric_cols[0]].dropna().values for name, group in df.groupby("regulation_status")]
    if len(groups) > 1:
        anova_result = f_oneway(*groups)
        print("\nüìä ANOVA: Diferencias en", numeric_cols[0], "seg√∫n regulaci√≥n")
        print("F-statistic:", anova_result.statistic, "| p-value:", anova_result.pvalue)
    else:
        print("\n‚ö†Ô∏è No hay suficientes categor√≠as en 'regulation_status' para ANOVA")
else:
    print("\n‚ö†Ô∏è No existe la columna 'regulation_status' en el dataset")

# ============================================
# üîπ 6. Feature Importance con RandomForest
# ============================================
if "ai_impact_index" in df.columns:
    X = df.drop(columns=["ai_impact_index", "cluster"], errors="ignore").select_dtypes(include=np.number)
    y = df["ai_impact_index"]

    rf = RandomForestRegressor(random_state=42)
    rf.fit(X, y)

    importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=True)

    plt.figure(figsize=(10,6))
    importances.plot(kind="barh", color="darkviolet")
    plt.title("Importancia de variables en AI Impact Index", fontsize=13, color="navy")
    plt.show()
else:
    print("\n‚ö†Ô∏è No existe la columna 'ai_impact_index' para calcular importancia de variables")

# ============================================
# üîπ 7. Relaci√≥n entre variables con regresi√≥n
# ============================================
if len(numeric_cols) >= 2:
    plt.figure(figsize=(8,6))
    sns.regplot(data=df, x=numeric_cols[0], y=numeric_cols[1],
                scatter_kws={"alpha":0.5}, line_kws={"color":"darkviolet"})
    plt.title(f"Relaci√≥n entre {numeric_cols[0]} y {numeric_cols[1]}", fontsize=13, color="navy")
    plt.show()

# ============================================
# üîπ 8. Ridgeline Plot (Joyplot) por a√±o
# ============================================
try:
    import joypy
    if "year" in df.columns and numeric_cols[0] in df.columns:
        plt.figure(figsize=(12,8))
        joypy.joyplot(df, by="year", column=numeric_cols[0],
                      colormap=sns.color_palette("mako", as_cmap=True),
                      fade=True, figsize=(12,8))
        plt.title(f"Distribuci√≥n de {numeric_cols[0]} a lo largo de los a√±os",
                  fontsize=14, color="darkviolet")
        plt.show()
    else:
        print("\n‚ö†Ô∏è No existe columna 'year' para hacer joyplot")
except ImportError:
    print("\n‚ö†Ô∏è Instalar joypy con: pip install joypy para habilitar el ridgeline plot")

# ============================================
# ‚úÖ Fin del EDA Final Avanzado
# ============================================
```

üìñ Descripci√≥n de los gr√°ficos
1Ô∏è‚É£ Distribuciones KDE (Kernel Density Estimation)

üìå Qu√© muestra:
La forma de las distribuciones de las primeras 5 variables num√©ricas.
üìå Interpretaci√≥n:

Picos = valores m√°s comunes.

Anchura = variabilidad de la variable.

Permite ver si las m√©tricas siguen patrones normales, sesgados o multimodales.

2Ô∏è‚É£ Pairplot Multivariado

üìå Qu√© muestra:
Matriz de gr√°ficos de dispersi√≥n y distribuciones entre 4 variables num√©ricas clave.
üìå Interpretaci√≥n:

Diagonal = distribuciones individuales de cada variable.

Fuera de la diagonal = relaciones entre pares de variables.

Se pueden detectar correlaciones lineales, agrupaciones o valores at√≠picos.

3Ô∏è‚É£ Clustering KMeans

üìå Qu√© muestra:
Un gr√°fico de dispersi√≥n (scatterplot) con colores seg√∫n los clusters creados por KMeans.
üìå Interpretaci√≥n:

Cada color = grupo de pa√≠ses/industrias con caracter√≠sticas similares.

Permite segmentar el dataset en perfiles, por ejemplo: alta adopci√≥n de IA y alta confianza, vs. baja adopci√≥n y bajo impacto.

4Ô∏è‚É£ ANOVA (An√°lisis de Varianza)

üìå Qu√© muestra (no es gr√°fico, es estad√≠stico):
Prueba si las medias de una m√©trica (ej. adopci√≥n de IA) son significativamente diferentes seg√∫n estatus de regulaci√≥n.
üìå Interpretaci√≥n:

p-value < 0.05 ‚Üí diferencias estad√≠sticamente significativas.

p-value > 0.05 ‚Üí no hay evidencia de diferencias notables.

5Ô∏è‚É£ Importancia de Variables (RandomForest)

üìå Qu√© muestra:
Un gr√°fico de barras con la importancia de cada variable num√©rica en la predicci√≥n de ai_impact_index.
üìå Interpretaci√≥n:

Barras m√°s largas = variables con m√°s peso en explicar el impacto de la IA.

Permite identificar drivers principales (ej: confianza en IA, ingresos por IA, colaboraci√≥n humano-IA).

6Ô∏è‚É£ Regresi√≥n Lineal (Regplot)

üìå Qu√© muestra:
Relaci√≥n entre las dos primeras variables num√©ricas, con l√≠nea de tendencia ajustada.
üìå Interpretaci√≥n:

L√≠nea ascendente ‚Üí relaci√≥n positiva (a mayor X, mayor Y).

L√≠nea descendente ‚Üí relaci√≥n negativa (a mayor X, menor Y).

Dispersi√≥n alrededor de la l√≠nea ‚Üí indica qu√© tan fuerte es la relaci√≥n.

7Ô∏è‚É£ Ridgeline Plot (Joyplot)

üìå Qu√© muestra:
Evoluci√≥n de la distribuci√≥n de una m√©trica num√©rica a lo largo de los a√±os.
üìå Interpretaci√≥n:

Cada curva = distribuci√≥n en un a√±o.

Permite ver cambios temporales, como crecimiento de adopci√≥n, confianza o impacto de la IA con el tiempo.

‚úÖ En conjunto, estos gr√°ficos permiten:

Explorar la distribuci√≥n de m√©tricas clave.

Detectar correlaciones y patrones multivariados.

Identificar segmentos en el dataset con clustering.

Ver diferencias estad√≠sticas entre grupos (ANOVA).

Reconocer las variables m√°s influyentes en el impacto de la IA.

Analizar tendencias temporales con joyplot.
